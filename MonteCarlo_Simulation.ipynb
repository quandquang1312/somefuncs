The idea behind estimating Pi number using MC simulation is the formula for area of square and circle.

Let $r$ is the side length of a square and also the radius of a circle.

We have:
$$\large{S_{circle} = \pi r^2}$$
$$\large{S_{square} = r^2}$$ 
=>
$$\large{\frac{s_{circle}}{s_{square}}} = \pi$$


```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

For a quarter of an $r$ radius circle. The area of it above the area of a square (with $r$ is side length) equals to $\large{\frac{1}{4}\pi}$.

All we need is to keep track the number of points inside the quarter and the total we throw randomly into the figure.


   $$\large{\frac{number\:of\:points\:inside}{number\:of\:points} = \frac{1}{4}\pi \implies \pi = 4 * (\frac{number\:of\:points\:inside}{number\:of\:points})}$$


   And for determining if a point $(x,y)$ is inside:
$$\large{x^2+y^2 \leq r^2}$$

```python
class MonteCarlo():
    def __init__(self):
        self.square_points = []
        self.circle_points = []
        self.pi = 0

    def estimating_pi(self, interval=1000):
    
        for i in range(interval ** 2):
            
            x = random.uniform(0,1)
            y = random.uniform(0,1)
            self.square_points.append((x,y))

            if x**2 + y**2 <= 1:
                self.circle_points.append((x,y))

        self.pi = 4 * len(self.circle_points) / len(self.square_points)
            
    def plot_simulation(self, no_points=369):
        points_inside = np.array(self.circle_points[:no_points])
        points = np.array(self.square_points[:no_points])

        plt.plot(points[:,0], points[:,1], 'go')
        plt.plot(points_inside[:,0], points_inside[:,1], 'ro')
        plt.title("Pi estimated: {}".format(self.pi))
        plt.axis('square')
        plt.show()
```

```python
simulation = MonteCarlo()
simulation.estimating_pi(5000)
simulation.plot_simulation()
```
