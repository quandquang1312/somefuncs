```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
class MonteCarlo():
    def __init__(self):
        self.square_points = []
        self.circle_points = []
        self.pi = 0

    def estimating_pi(self, interval=1000):
    
        for i in range(interval ** 2):
            
            x = random.uniform(0,1)
            y = random.uniform(0,1)
            self.square_points.append((x,y))

            if x**2 + y**2 <= 1:
                self.circle_points.append((x,y))

        self.pi = 4 * len(self.circle_points) / len(self.square_points)
            
    def plot_simulation(self, no_points=369):
        points_inside = np.array(self.circle_points[:no_points])
        points = np.array(self.square_points[:no_points])

        plt.plot(points[:,0], points[:,1], 'go')
        plt.plot(points_inside[:,0], points_inside[:,1], 'ro')
        plt.title("Pi estimated: {}".format(self.pi))
        plt.axis('square')
        plt.show()
```

```python
simulation = MonteCarlo()
simulation.estimating_pi(5000)
simulation.plot_simulation()
```
